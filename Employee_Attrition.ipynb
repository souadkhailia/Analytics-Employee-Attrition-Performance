```python
plt.figure(figsize=(8, 6))
sns.histplot(df['Age'], bins=20, kde=True)
plt.title('Distribution of Age')
plt.xlabel('Age')
plt.ylabel('Frequency')
plt.show()
# Interpretation: The histogram shows the distribution of ages among employees. The peak in the middle indicates that most employees are in their 30s.
```

```python


# Create a countplot to visualize attrition rates
plt.figure(figsize=(8, 6))
sns.countplot(x='Attrition', data=df, palette='Set1')
plt.title('Attrition Overview')
plt.xlabel('Attrition')
plt.ylabel('Count')
plt.show()
#Employee leaving the company (no or yes)

```

```python

sns.set(style="whitegrid")
plt.figure(figsize=(8, 6))
sns.countplot(x="Attrition", hue="Gender", data=df)

plt.xlabel("Attrition")
plt.ylabel("Count")
plt.title("Attrition Rates by Gender")

# Show the plot
plt.legend(title="Attrition", loc="upper right")
plt.show()
```

```python

# Create a scatter plot of age vs. years at company
plt.scatter(df['Age'], df['YearsAtCompany'])
plt.title('Scatter Plot of Age vs. Years AtCompany')
plt.xlabel('Age')
plt.ylabel('Years AtCompany')

# Display the plot
plt.show()

# Interpretation: The scatter plot shows the relationship between age and years at company. There is a general upward trend, indicating that employees tend to have more years at company as they get older.
```

```python
# Define the mapping of values to labels
job_satisfaction_labels = {1: 'Low', 2: 'Medium', 3: 'High', 4: 'Very High'}

# Map the integer values to labels
df['JobSatisfaction'] = df['JobSatisfaction'].map(job_satisfaction_labels)

plt.figure(figsize=(30, 25))
subset_data = df[['Age', 'TotalWorkingYears', 'JobSatisfaction']]
sns.pairplot(subset_data, hue='JobSatisfaction', palette='Set1')
#plt.title('Pairplot of Age, TotalWorkingYears, and JobSatisfaction')
plt.show()


```

```python


# Define the mapping of values to labels
work_life_balance_labels = {1: 'Bad', 2: 'Good', 3: 'Better', 4: 'Best'}
relationship_satisfaction_labels = {1: 'Low', 2: 'Medium', 3: 'High', 4: 'Very High'}
performance_rating_labels = {1: 'Low', 2: 'Good', 3: 'Excellent', 4: 'Outstanding'}

# Map the integer values to labels
df['WorkLifeBalance'] = df['WorkLifeBalance'].map(work_life_balance_labels)
df['RelationshipSatisfaction'] = df['RelationshipSatisfaction'].map(relationship_satisfaction_labels)
df['PerformanceRating'] = df['PerformanceRating'].map(performance_rating_labels)
# Create the first heatmap
plt.figure(figsize=(8, 6))
heatmap_data1 = pd.crosstab(df['WorkLifeBalance'], df['RelationshipSatisfaction'])
sns.heatmap(heatmap_data1, annot=True, fmt='d', cmap='coolwarm', cbar=True)
plt.xlabel("Relationship Satisfaction")
plt.ylabel("Work-life Balance")
plt.title("Relationship Between Work-life Balance and Relationship Satisfaction")

# Create the second heatmap
plt.figure(figsize=(8, 6))
heatmap_data2 = pd.crosstab(df['PerformanceRating'], df['RelationshipSatisfaction'])
sns.heatmap(heatmap_data2, annot=True, fmt='d', cmap='coolwarm', cbar=True)
plt.xlabel("Relationship Satisfaction")
plt.ylabel("Performance Rating")
plt.title("Relationship Between Performance Rating and Relationship Satisfaction")

# Show the plots
plt.show()
```
